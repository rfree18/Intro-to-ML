```python
import numpy
from pprint import pprint as pp
```

```python
from sklearn import datasets
```

```python
newsgroups_train = datasets.fetch_20newsgroups(subset='train', 
                                               categories=['comp.sys.mac.hardware', 'comp.sys.ibm.pc.hardware'])
```

```python
newsgroups_test = datasets.fetch_20newsgroups(subset='test', 
                                               categories=['comp.sys.mac.hardware', 'comp.sys.ibm.pc.hardware'])
```

```python
print newsgroups_train.target[0], newsgroups_train.data[0]
```

```python
print newsgroups_train.target[2], newsgroups_train.data[2]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
vectorizer = CountVectorizer(min_df=30)
vectors = vectorizer.fit_transform(newsgroups_train.data)
vectors_test = vectorizer.transform(newsgroups_test.data)
```

```python
pp(vectorizer.vocabulary_)
```

```python
print vectors[0]
```

```python
print vectors[3]
```

```python
from sklearn.linear_model import Perceptron
perceptron = Perceptron()
```

```python
perceptron = perceptron.fit(vectors, newsgroups_train.target)
```

```python
y_train = perceptron.predict(vectors)
y_test = perceptron.predict(vectors_test)
```

```python
float(numpy.sum(abs(y_train - newsgroups_train.target)))/len(newsgroups_train.target)
```

```python
float(numpy.sum(abs(y_test - newsgroups_test.target)))/len(newsgroups_test.target)
```

```python
invocab = dict([(vv,kk) for kk, vv in vectorizer.vocabulary_.items()])
```

```python
insens_idx = numpy.argsort(abs(perceptron.coef_.squeeze()))
ordered_idx = numpy.argsort(perceptron.coef_.squeeze())
```

```python
print 'Insensitive to:'
for jj in insens_idx[:10]:
    print '{} ({:0.2f})'.format(invocab[jj], perceptron.coef_[0,jj])
print

print 'Sensitive to:'
for jj in insens_idx[-10:]:
    print '{} ({:0.2f})'.format(invocab[jj], perceptron.coef_[0,jj])
print
```

```python
print 'Favours ibm:'
for jj in ordered_idx[:10]:
    print '{} ({:0.2f})'.format(invocab[jj], perceptron.coef_[0,jj])
print

print 'Favours mac:'
for jj in ordered_idx[-10:]:
    print '{} ({:0.2f})'.format(invocab[jj], perceptron.coef_[0,jj])
print
```
