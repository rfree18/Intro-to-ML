```python
import numpy
from scipy.stats import *
```

```python
from matplotlib import pyplot as plot
matplotlib.rc('text', usetex=True)
```

```python
rv = gamma(a=2.3)

x_min = 0
x_max = 5
x = numpy.linspace(x_min, x_max, 100)

p_x = rv.pdf(x)

F_x = rv.cdf(x)

mean, var = rv.stats(moments='mv')
mode = x[p_x.argmax()]

plot.figure()

plot.plot(x, p_x, 'r-')
plot.plot(x, F_x, 'b-')
plot.plot([mean, mean], [0, 1], 'b--')
plot.plot([mode, mode], [0, 1], 'r--')

plot.xlim([x_min, x_max])
plot.ylim([0, 1])

plot.grid('on')
plot.legend(['PDF', 'CDF', 'Mean', 'Mode'], loc='best')
plot.title('Gamma(2.3)')

plot.show()
plot.tight_layout()
```

```python
plot.savefig('../figures/gamma.pdf', dpi=100, bbox_inches='tight')
```

```python
rv = norm(2,1)

x_min = 0
x_max = 5
x = numpy.linspace(x_min, x_max, 100)

p_x = rv.pdf(x)

F_x = rv.cdf(x)

mean, var = rv.stats(moments='mv')
mode = x[p_x.argmax()]

plot.figure()

plot.plot(x, p_x, 'r-')
plot.plot(x, F_x, 'b-')
plot.plot([mean, mean], [0, 1], 'b--')
plot.plot([mode, mode], [0, 1], 'r--')
x_ = numpy.linspace(mean-numpy.sqrt(var), mean+numpy.sqrt(var), 100)
plot.fill_between(x_, rv.pdf(x_), alpha=0.4)

plot.xlim([x_min, x_max])
plot.ylim([0, 1])

plot.grid('on')
plot.legend(['PDF', 'CDF', 'Mean', 'Mode', '+- Std. Dev.'], loc='best')
plot.title('Normal(2,1)')

plot.show()
plot.tight_layout()
```

```python
plot.savefig('../figures/gauss.pdf', dpi=100, bbox_inches='tight')
```

```python
rv = multivariate_normal(numpy.array([1.5, 0.5]), numpy.array([[3., 1.], [0.5, 2]]))
```

```python
x_ = numpy.linspace(-3, 4, 100)
y_ = numpy.linspace(-3, 4, 100)
xx, yy = numpy.meshgrid(x_, y_)

probs = rv.pdf(numpy.concatenate([xx.ravel()[:,None], yy.ravel()[:,None]], axis=1))

plot.figure()
plot.contourf(xx, yy, probs.reshape(xx.shape), cmap=plot.cm.Blues, alpha=1.)
plot.grid('on')
plot.xlabel('$X$')
plot.ylabel('$Y$')
plot.show()
plot.tight_layout()
plot.savefig('../figures/joint.pdf', dpi=100, bbox_inches='tight')
```

```python
x_ = numpy.linspace(-3, 4, 100)

plot.figure()
legends = []
for yy in [-2, -1, 0, 1, 2]:
    y_ = numpy.zeros(100) + yy
    probs = rv.pdf(numpy.concatenate([x_[:,None], y_[:,None]], axis=1))
    plot.plot(x_, probs)
    legends.append('$p(X|Y={})$'.format(yy))

plot.grid('on')
plot.xlim([-3, 4])
plot.ylim([0, 0.07])
plot.legend(legends, loc='best')
plot.show()
plot.tight_layout()
plot.savefig('../figures/conditionals.pdf', dpi=100, bbox_inches='tight')
```
