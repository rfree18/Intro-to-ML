```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.decomposition import PCA
from sklearn.datasets import fetch_mldata
```

## 1. Data

```python
mnist = fetch_mldata('MNIST original')
```

```python
X = mnist.data
y = mnist.target
print ('X Shape: ', X.shape)
```

```python
plt.imshow(X[100].reshape((28, 28)), cmap=cm.gray)
```

## 2. PCA - Principal Components Analysis

```python
# Transforming the data

pca = PCA(n_components=2)

X_reduced = pca.fit_transform(X)
```

```python
X_reduced.shape
```

## 3. Visualizing reduced data points

```python
plt.figure(figsize=(17, 15))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c = y)
plt.colorbar()
```

## 4. Analysis

```python
# Important thing - How much variance is explained by each dimension

print ('Ratio of variance explained: ', pca.explained_variance_ratio_)
```

## 5. Visualizing the principal components

```python
principal_components = pca.components_
```

```python
principal_components.shape
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.imshow(principal_components[0].reshape((28, 28)), cmap=cm.gray)

plt.subplot(1, 3, 3)
plt.imshow(principal_components[1].reshape((28, 28)), cmap=cm.gray)
```

## 6. Reconstructing an image

```python
pca_rec = PCA(n_components=200)

X_reduced_rec = pca_rec.fit_transform(X)
```

```python
principal_components_rec = pca_rec.components_
```

```python
plt.imshow(X[18000].reshape((28, 28)), cmap=cm.gray)
```

```python
rec = np.dot(X_reduced_rec[18000], principal_components_rec)
```

```python
plt.imshow(rec.reshape((28, 28)), cmap=cm.gray)
```
