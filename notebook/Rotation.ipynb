```python
%matplotlib notebook 

import numpy
import matplotlib.pyplot as plot
```

```python
import matplotlib
matplotlib.rc('text', usetex=True)
```

```python
numpy.random.seed(1234)
```

```python
def label_map(y_, from_, to_):
    y = numpy.copy(y_)
    for f in from_:
        y[y_ == f] = to_
    return y
```

```python
from sklearn.datasets import make_blobs
n_dim = 2
x_tra, y_tra = make_blobs(n_samples=100, n_features=n_dim, centers=[[1,1],[-1,-1],[1,-1],[-1,1]], shuffle=True, cluster_std=0.3)

y_tra = label_map(y_tra, [0, 1], 0)
y_tra = label_map(y_tra, [2, 3], 1)
```

```python
theta = numpy.radians(45)
c, s = numpy.cos(theta), numpy.sin(theta)
R = numpy.array([[c, -s], [s, c]])
```

```python
x_r = numpy.dot(x_tra, R)
x_ra = abs(x_r)
```

```python
# visualize data 
def vis_data(x, y, c='r'):
    plot.hold('on')
    for x_, y_ in zip(x, y):
        plot.plot(x_[0], x_[1], c+'o' if y_ == 0 else c+'+')
    plot.hold('off')
    plot.grid('on')
```

```python
plot.figure()

plot.subplot(1,3,1)
vis_data(x_tra, y_tra, c='r')
plot.title('XOR (Original)')
plot.tight_layout()

plot.subplot(1,3,2)
vis_data(x_r, y_tra, c='r')
plot.title('XOR (Rotated)')
plot.tight_layout()

plot.subplot(1,3,3)
vis_data(x_ra, y_tra, c='r')
plot.title('XOR (Rotated+Abs)')
plot.tight_layout()

plot.tight_layout()
plot.show()
```

```python
plot.savefig('../figures/xor_transform.pdf', dpi=100, bbox_inches='tight')
```
